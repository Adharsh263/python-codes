```python
# importing pandas module
import pandas as pd

# importing numpy as np
import numpy as np

# Define a dictionary containing employee data
data1 = {'Name':['Jai', 'Anuj', 'Jai', 'Princi',
                 'Gaurav', 'Anuj', 'Princi', 'Abhi'],
        'Age':[27, 24, 22, 32,
               33, 36, 27, 32],
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj',
                   'Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd','B.Tech', 'B.com', 'Msc', 'MA'],
        'Score': [23, 34, 35, 45, 47, 50, 52, 53]}


# Convert the dictionary into DataFrame
df = pd.DataFrame(data1)
display(df)
                         
                         
                     
```

```python
import numpy as pd
name_group=df.groupby('Name')
name_group.aggregate(np.sum)
```

```python
group2=df.groupby(['Name','Qualification'])
group2.aggregate(np.sum)
```

```python
name_group=df.groupby('Name')
name_group['Age'].agg([np.sum,np.mean,np.std])
```

```python
name_group=df.groupby('Name')
name_group.agg({'Age':'sum','Score':'mean'})
```

```python
name_group=df.groupby('Name')
sc=lambda x:(x - x.mean()) / x.std()*10
name_group[['Age','Score']].transform(sc)
```

```python

data = {'Team': ['A', 'A', 'B', 'B', 'C', 'C'],
        'Score': [10, 20, 15, 30, 50, 60]}
df=pd.DataFrame(data)
display(df)
```

```python
teamgroup=df.groupby('Team')
df['mean']=teamgroup[['Score']].transform("mean")
display(df)
```
