```python
import pandas as pd
df=pd.read_csv("C:/Users/adhar/OneDrive/Documents/python/nba (1).csv")
df=df.drop(457)
def convert_height(height):
    feet,inches=map(int,height.split('-'))
    return feet*.3048 + inches* .0254
df=df.assign(height_in_meter=df['Height'].apply(convert_height))
df=df.assign(BMI=lambda x:x['Weight']/(x['height_in_meter']**2))
display(df)
```
